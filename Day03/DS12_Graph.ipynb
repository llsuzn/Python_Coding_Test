## 그래프

```python
class Graph:
    def __init__(self,size) -> None:
        self.SIZE = size
        self.graph = [[0 for _ in range(size)] for _ in range(size)] # 2차원 배열

G1 = Graph(4) # 4행 4열

for item in G1.graph: # 2차원 배열 형태로 출력
    print(item)    
```

가중치 X 무방향 그래프



```python
# A 출발
G1.graph[0][1] = 1 # (A,B) edge(간선)
G1.graph[0][2] = 1 # (A,C) edge(간선)
G1.graph[0][3] = 1 # (A,D) edge(간선)
# B 출발
G1.graph[1][0] = 1 # (B,A) edge(간선)
G1.graph[1][2] = 1 # (B,C) edge(간선)
# C 출발
G1.graph[2][0] = 1 # (C,A) edge(간선)
G1.graph[2][1] = 1 # (C,B) edge(간선)
G1.graph[2][3] = 1 # (C,D) edge(간선)
# D 출발
G1.graph[3][0] = 1 # (D,A) edge(간선)
G1.graph[3][2] = 1 # (D,C) edge(간선)

print('무방향 그래프')
for item in G1.graph: # 2차원 배열 형태로 출력
    print(item) 
```

가중치 X 방향성 그래프



```python
# 방향성 그래프

G3 = Graph(4)

# A
G3.graph[0][1] = 1
G3.graph[0][2] = 1
# D
G3.graph[3][0] = 1
G3.graph[3][2] = 1

print('방향성 그래프')
for item in G3.graph: # 2차원 배열 형태로 출력
    print(item) 
```
